**False** - The false value of the bool type. Assignments to False are illegal and raise a SyntaxError.<br>
**True** - The true value of the bool type. Assignments to True are illegal and raise a SyntaxError.<br>
**None** - An object frequently used to represent the absence of a value, as when default arguments are not passed to a function. Assignments to None are illegal and raise a SyntaxError. None is the sole instance of the NoneType type.<br>
**NotImplemented** - A special value which should be returned by the binary special methods<br>
**Ellipsis** - The same as the ellipsis literal “...”. Special value used mostly in conjunction with extended slicing syntax for user-defined container data types. Ellipsis is the sole instance of the types.EllipsisType type.<br>
**__debug__** - This constant is true if Python was not started with an -O option<br>
